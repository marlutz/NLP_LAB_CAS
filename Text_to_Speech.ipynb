## Voice Clone

To clone a voice, you need to provide a reference audio clip (ref_audio) with its transcript (ref_text). The audio clip can be a local file path, a URL (or a base64 string, or a (numpy_array, sample_rate) tuple). If you set x_vector_only_mode=True, ref_text is not required (only the speaker embedding is used) but cloning quality may be reduced.

```python
model = Qwen3TTSModel.from_pretrained(
    "Qwen/Qwen3-TTS-12Hz-1.7B-Base",
    #"Qwen/Qwen3-TTS-12Hz-0.6B-Base",
    device_map="auto",
    dtype=torch.bfloat16,
    #attn_implementation="flash_attention_2",
)
```

```python
cloned_voice = "data/clone.wav"
download("https://drive.switch.ch/index.php/s/TvH5YScf7XyB45y/download", cloned_voice)
```

```python
#ref_audio = "https://qianwen-res.oss-cn-beijing.aliyuncs.com/Qwen3-TTS-Repo/clone.wav"
#ref_text  = "Okay. Yeah. I resent you. I love you. I respect you. But you know what? You blew it! And thanks to you."
ref_audio = cloned_voice
ref_text  = "Of course, the answer to the slavery question was already embedded within our constitution."
```

```python
wavs, sr = model.generate_voice_clone(
    text="I really like the NLP Lab!",
    language="English",
    ref_audio=ref_audio,
    ref_text=ref_text,
)
```

```python
pathlib.Path("data").mkdir(parents=True, exist_ok=True)

for i, w in enumerate(wavs):
    IPython.display.display(Audio(data=w, rate=sr, embed=True))
    sf.write(f"data/output_voice_clone_{i}.wav", w, sr)
```

If you want to reuse the same voice for multiple generations (to avoid recomputation), build it once with `create_voice_clone_prompt` and then pass it to `voice_clone_prompt`.

```python
prompt_items = model.create_voice_clone_prompt(
    ref_audio=ref_audio,
    ref_text=ref_text,
    x_vector_only_mode=False,
)
```

```python
wavs, sr = model.generate_voice_clone(
    text=[
      "Ich werde meine Hausaufgaben ab jetzt stets gewissenhaft erledigen!",
      "From now on, I will always do my homework conscientiously!"
    ],
    language=["German", "English"],
    voice_clone_prompt=prompt_items,
)
```

```python
pathlib.Path("data").mkdir(parents=True, exist_ok=True)

for i, w in enumerate(wavs):
    IPython.display.display(Audio(data=w, rate=sr, embed=True))
    sf.write(f"data/output_voice_clone_reuse_{i}.wav", w, sr)
```

## Voice Design then Clone

You can use following workflow to designe a reusable voice:
1. use the `VoiceDesign` model to synthesize a short reference clip that matches your target persona
2. feed that clip into `create_voice_clone_prompt` to build a reusable voice
3. call `generate_voice_clone` with `voice_clone_prompt` to generate new content

This is useful when you want a consistent character/voice.

```python
# create a reference audio in the target style using the VoiceDesign model

design_model = Qwen3TTSModel.from_pretrained(
    "Qwen/Qwen3-TTS-12Hz-1.7B-VoiceDesign",
    #"Qwen/Qwen3-TTS-12Hz-0.6B-VoiceDesign",
    device_map="auto",
    dtype=torch.bfloat16,
    #attn_implementation="flash_attention_2",
)
```

```python
ref_text = "H-hey! You dropped your... uh... calculus notebook? I mean, I think it's yours? Maybe?"
ref_instruct = "Male, 17 years old, tenor range, gaining confidence - deeper breath support now, though vowels still tighten when nervous"

ref_wavs, sr = design_model.generate_voice_design(
    text=ref_text,
    language="English",
    instruct=ref_instruct
)
```

```python
pathlib.Path("data").mkdir(parents=True, exist_ok=True)

IPython.display.display(Audio(data=ref_wavs[0], rate=sr, embed=True))
sf.write("data/voice_design_reference.wav", ref_wavs[0], sr)
```

```python
# build a reusable clone prompt from the voice design reference

clone_model = Qwen3TTSModel.from_pretrained(
    "Qwen/Qwen3-TTS-12Hz-1.7B-Base",
    #"Qwen/Qwen3-TTS-12Hz-0.6B-Base",
    device_map="auto",
    dtype=torch.bfloat16,
    #attn_implementation="flash_attention_2",
)
```

```python
voice_clone_prompt = clone_model.create_voice_clone_prompt(
    ref_audio=(ref_wavs[0], sr),   # or "data/voice_design_reference.wav"
    ref_text=ref_text,
)
```

```python
sentences = [
    "No problem! I actually... kinda finished those already? If you want to compare answers or something...",
    "What? No! I mean yes but not like... I just think you're... your technique is really precise!",
]
```

```python
wavs, sr = clone_model.generate_voice_clone(
    text=sentences,
    language=["English", "English"],
    voice_clone_prompt=voice_clone_prompt,
)
```

```python
pathlib.Path("data").mkdir(parents=True, exist_ok=True)

for i, w in enumerate(wavs):
    IPython.display.display(Audio(data=w, rate=sr, embed=True))
    sf.write(f"data/output_voice_design_clone_reuse_{i}.wav", w, sr)
```

Interactive UI to generate Speech... (only works on Colab)

```python
# @title 🎙️ Qwen3-TTS Interactive Generator { display-mode: "form" }

# @markdown ---
# @markdown ### Input Settings
text_to_synthesize = "I can't believe I finally finished this project!" # @param {type:"string"}
language = "English" # @param ["English", "German"]
# @markdown ---

def generate_interactive_speech(text, lang):
    if not text.strip():
        print("Please enter some text.")
        return

    print(f"🎬 Generating speech in {lang}...")
    # We use the 'clone_model' and 'voice_clone_prompt'
    # that you initialized in your previous cells.
    try:
        wavs, sr = clone_model.generate_voice_clone(
            text=[text],
            language=[lang],
            voice_clone_prompt=voice_clone_prompt,
        )

        # Display the audio player immediately
        from IPython.display import Audio, display
        for w in wavs:
            display(Audio(data=w, rate=sr, embed=True))
        print("✅ Done!")
    except NameError:
        print("❌ Error: 'clone_model' or 'voice_clone_prompt' not found.")
        print("Please make sure you have run the previous cells first!")

# Run the function with the form inputs
generate_interactive_speech(text_to_synthesize, language)
```
